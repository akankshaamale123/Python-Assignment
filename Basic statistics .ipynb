```python
sales_data.boxplot(figsize = (10,8))
plt.suptitle("Boxplot of numerical cols")
plt.show()
```

```python
"""
Box represent the middle 50% of the data in interquartile Range : q1 to q3
IQR = upper quartile (q3) - lower quartile(q1)
Upper extreme = UQ + 1.5 * IQR
LowerExtreme = LQ - 1.5 * IQR

point outside srepresent outliers.
A tall box suggest high variability in the data.
A short box suggest low variability in the data.
"""
```
